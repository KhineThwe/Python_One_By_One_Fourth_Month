## Temp Humidity BarChart

## Cus DataSet

```python
import numpy as np
import pandas as pd
from collections import Counter
plt.style.use("fivethirtyeight")
data = pd.read_csv("sample product.csv")
print(data.head(5))

customers = data['Customer']
products = data['Product']

customer_count = Counter()

# myIter = ['x','y','z','x','x','y','z','z','z','z']
# print(Counter(myIter))

for cus in customers:
    customer_count.update((cus))
print(customer_count)

most_cus = []
rate = []

for i in customer_count.most_common(10):
    most_cus.append(i[0])
    rate.append(i[1])
print(most_cus,rate)

most_cus.reverse()
rate.reverse()
plt.barh(most_cus,rate)
plt.ylabel("Most_Cus")
plt.xlabel("Rate")
plt.title("Most Cus & Rate")
plt.tight_layout()
plt.savefig("MostCus.png")
plt.show()
```

## Pie Chart

```python
#1
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
# left

right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]
# right

labels = ['Left','Right']
colors = ["#abcdef","#aabbcc"]
plt.pie([left,right],labels=labels,colors=colors,autopct='%.2f%%')
plt.title("Foot Preference of FIFA Players")
plt.show()
```

## Pie Chart #2

```python
#simple
fifa.Weight = [int(x.strip('lbs')) if type(x)==str else x for x in fifa.Weight]
fifa.Weight#lbs eliminated


light = fifa.loc[fifa.Weight < 125 ].count()[0]
light


light_medium = fifa[(fifa.Weight >= 125) & (fifa.Weight < 150)].count()[0]
light_medium

medium = fifa[(fifa.Weight >= 150) & (fifa.Weight < 175)].count()[0]
medium


medium_heavy =  fifa[(fifa.Weight >= 175) & (fifa.Weight < 200)].count()[0]
medium_heavy


heavy = fifa[(fifa.Weight>=200)].count()[0]
heavy


#2
plt.figure(figsize=(8,5),dpi=300)

plt.style.use("ggplot")
weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ['under 125','125-150','150-175','175-200','over 200']
explodevalue = (.4,.2,0,0,.4)
plt.title("Weight of Professional Soccer Player (lbs)")
plt.pie(weights,labels=labels,explode=explodevalue,pctdistance=0.8,autopct='%.2f%%')
plt.show()
```

## Box and Whiskers Chart

```python
plt.figure(figsize=(5,8),dpi=100)
plt.style.use('default')

barcelona = fifa.loc[fifa.Club == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa.Club == 'Real Madrid']['Overall']
revs = fifa.loc[fifa.Club == "New England Revolution"]['Overall']

#plt.boxplot(data,labels)
bp = plt.boxplot([barcelona,madrid,revs],labels=['FC Barcelona','Real Madrid','NE Revolution'],patch_artist=True,medianprops={'linewidth':2})

plt.title("Professional Soccer Team Comparison")
plt.ylabel('FIFA Overall Rating')

for box in bp['boxes']:
    #change outline color
    box.set(color='#4286f4',linewidth=2)
    # change fill color
    box.set(facecolor = '#e0e0e0' )
#     #change hatch
#     box.set(hatch='/')
    
plt.show()
```
